### Descripción del problema

#### El concepto central es la Programación Orientada a Objetos (POO) aplicada a la gestión de eventos. En este modelo, la clase Evento actúa como una plantilla que define las características necesarias (atributos: nombre, fecha, lugar y capacidad) y los comportamientos posibles (funciones: registrar, invitar y generar entradas). Al ejecutar el código, se crean dos objetos independientes (evento1 y evento2) que, mediante la interacción directa con el usuario vía consola, cobran vida con datos únicos, permitiendo que cada instancia procese su propia información y ejecute sus acciones de forma autónoma.

```python
class Evento:
    def __init__(self, nombre, fecha, lugar, capacidad_maxima):
        # Atributos de la clase
        self.nombre = nombre
        self.fecha = fecha
        self.lugar = lugar
        self.capacidad_maxima = capacidad_maxima

    # Funciones (Métodos) de la clase
    def registrar_asistente(self):
        print(f"✅ ¡Registro exitoso! Te has anotado para el evento: '{self.nombre}'.")

    def enviar_invitacion(self):
        print(f"📧 Enviando invitación para el día {self.fecha} en {self.lugar}...")

    def generar_entrada(self):
        print(f"🎟️ Generando ticket QR... Capacidad restante: {self.capacidad_maxima - 1} espacios.")

def solicitar_datos_evento(numero):
    print(f"\n--- Ingrese los datos para el Evento {numero} ---")
    nombre = input("Nombre del evento: ")
    fecha = input("Fecha (ej. 15 de Octubre): ")
    lugar = input("Lugar: ")
    capacidad = int(input("Capacidad máxima de personas: "))

    # Retornamos el objeto creado a partir de la clase
    return Evento(nombre, fecha, lugar, capacidad)

# --- PROGRAMA PRINCIPAL ---

# 1. Creamos los 2 objetos pidiendo datos al usuario
evento1 = solicitar_datos_evento(1)
evento2 = solicitar_datos_evento(2)

# 2. Interactuamos con el objeto evento1
print(f"\n>>> Ejecutando acciones para: {evento1.nombre}")
evento1.enviar_invitacion()
evento1.registrar_asistente()
evento1.generar_entrada()

# 3. Interactuamos con el objeto evento2
print(f"\n>>> Ejecutando acciones para: {evento2.nombre}")
evento2.enviar_invitacion()
evento2.registrar_asistente()
evento2.generar_entrada()

print("\n--- Proceso finalizado con éxito ---")
print("Programa realizado por Angela Michelle Correa Lopez NL = 13")
```
